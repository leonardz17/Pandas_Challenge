```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "../Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#get total count of unique SNs 
total_players = len(purchase_data["SN"].unique())
totalplayers_df = pd.DataFrame({"Total Players":[total_players]})
totalplayers_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#run calculations 
item_count = len(purchase_data["Item Name"].unique())
avg_price = purchase_data["Price"].mean()
total_purchases = purchase_data["Purchase ID"].count()
revenue = purchase_data["Price"].sum()

#summary DataFrame
summary_df = pd.DataFrame({"Number of Unique Items": [item_count], 
                           "Average Price":avg_price,
                           "Number of Purchases":[total_purchases],
                           "Total Revenue":revenue})
summary_df.style.format({'Average Price':"${0:,.2f}",'Total Revenue':"${0:,.2f}"})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#locating data based on SN and Gender information, dropping information for duplicate SNs
gender_df = purchase_data.loc[:,["SN","Gender"]]
gender_reduced = gender_df.drop_duplicates(subset=["SN"])


#summary DataFrame
gender_count = gender_reduced["Gender"].value_counts()
percent_gender = gender_count/total_players * 100

summary_percent = pd.DataFrame({"Total Count": gender_count,
                               "Percentage of Players":percent_gender})
summary_percent.style.format({"Percentage of Players": "{:.2f}%"})

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#setting tables needed to extract values
gender_purchase = purchase_data.loc[:,["Gender","Purchase ID","Price"]]
purchase_count = gender_purchase["Gender"].value_counts()
grouped_purchases = gender_purchase.groupby("Gender")

#Calculating values
sum_price_gender = grouped_purchases["Price"].sum()
avg_price_gender = sum_price_gender/purchase_count
avg_per_person = sum_price_gender/gender_count

#summary DataFrame
summary_purchase = pd.DataFrame ({"Purchase Count":purchase_count, 
                                 "Average Purchase Price":avg_price_gender,
                                 "Total Purchase Value":sum_price_gender,
                                 "Average Purchase Price Per Person":avg_per_person})

summary_purchase.style.format({"Average Purchase Price":"${:.2f}",
                              "Total Purchase Value": "${:,.2f}",
                              "Average Purchase Price Per Person":"${:.2f}"})
```

* Create a summary data frame to hold the results

```python
#Take out duplicate players 
age_df = purchase_data.loc[:,["SN","Age"]]
# age_df_unique = age_df.drop_duplicates(subset=["SN"])


#create bins for ages of players
bins = [0,9.9,14,19,24,29,34,39,100]
group_labels = [
    "<10",
    "10-14",
    "15-19",
    "20-24",
    "25-29",
    "30-34",
    "35-39",
    "40+"]

#Placed data series into new column in DataFrame
age_df["Age Group"] = pd.cut(age_df["Age"], bins, labels=group_labels, include_lowest =True)
purchase_data

age_df_unique = age_df.drop_duplicates(subset=["SN"])
age_df_unique
```

```python
#Calculate count of age groups and percentage of players in each group
age_count = age_df_unique["Age Group"].value_counts()
percentage_age = age_count/total_players *100

#summary DataFrame
age_demo = pd.DataFrame({
    "Total Count":age_count,
    "Percentage of Players":percentage_age})
age_demo.style.format({"Percentage of Players": "{:.2f}%"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#data for SN and price for easy view
user_purchase = purchase_data.loc[:,["Purchase ID","SN","Price"]]
user_grouped = user_purchase.groupby("SN")
print(user_grouped)
user_grouped.count()
```

```python
#calculate purchase count by user
user_pcount = user_grouped["Purchase ID"].count()
sum_userp = user_grouped["Price"].sum()
avg_userp = sum_userp/user_pcount

#summary DataFrame
summary_userp = pd.DataFrame({
    "Purchase Count":user_pcount,
    "Average Purchase Price":avg_userp,
    "Total Purchase Value":sum_userp})
sorted_summary_userp = summary_userp.sort_values(["Purchase Count"],ascending = False)
sorted_summary_userp["Average Purchase Price"] = sorted_summary_userp["Average Purchase Price"].map("${:.2f}".format)
sorted_summary_userp["Total Purchase Value"] = sorted_summary_userp["Total Purchase Value"].map("${:.2f}".format)
sorted_summary_userp.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve Item ID, Item Name and Price
item_df = purchase_data.loc[:,["Item ID","Item Name","Price"]]
grouped_item = item_df.groupby(["Item ID","Item Name"])
grouped_item.count()
```

```python
#Calculate Purchase count,item price and total value
item_counts = grouped_item["Item Name"].count()
tprice_item = grouped_item["Price"].sum()
item_price = tprice_item/item_counts

#summary DataFrame
summary_items = pd.DataFrame({
    "Purchase Count":item_counts,
    "Item Price":item_price,
    "Total Purchase Value":tprice_item
})

sorted_summary_items = summary_items.sort_values("Purchase Count",ascending=False)
sorted_summary_items["Item Price"] = sorted_summary_items["Item Price"].map("${:.2f}".format)
sorted_summary_items["Total Purchase Value"] = sorted_summary_items["Total Purchase Value"].map("${:.2f}".format)
sorted_summary_items.head()
```
